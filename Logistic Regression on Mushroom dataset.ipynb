```python
#Logical Regression on mushroom data set
import numpy as np
import pandas as pd
from sklearn import linear_model
import os
```

```python
os.chdir("D:\\GEGDC\\KT00475829\\R work\\Data sets")
df = pd.read_csv("mushrooms.csv")
type(df)
```

```python
#Finding null values
df.isna().sum()
```

```python
df.describe()
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
lab_enc = LabelEncoder()
#Applying Label Encoding for the columns with no of classes = 2
#and applying One-hot enccoding for the rest of columns
for col in df.columns:
    if len(df[col].value_counts()) == 2:
        df[col] = lab_enc.fit_transform(df[col])

df = pd.get_dummies(df)
df.head()
    
```

```python
#separating out Target variable
y = df['class']
#Deleting Target from Dataset
df.drop('class', axis = 1)
df.head()
```

```python
y
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size = 0.2, random_state = 42)
```

```python
y_test
```

```python
#Building model
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
```

```python
#Predicting
y_pred = log_model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```
